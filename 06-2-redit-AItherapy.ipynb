```python
import requests
import pandas as pd

# ✅ Reddit JSON API URL
url = "https://www.reddit.com/r/evilautism/comments/1eoaep4/beginning_to_think_ai_therapy_might_not_be_a_bad/.json"

# ✅ Set user-agent
headers = {'User-Agent': 'Mozilla/5.0'}

# ✅ Fetch the data
response = requests.get(url, headers=headers)
if response.status_code != 200:
    print("❌ Failed to retrieve data:", response.status_code)
    exit()

# ✅ Parse the JSON
data = response.json()
comments = data[1]['data']['children']

# ✅ Store results
results = []

for comment in comments:
    if comment['kind'] != 't1':
        continue

    author = comment['data'].get('author', 'Unknown')
    content = comment['data'].get('body', 'No content')
    upvotes = comment['data'].get('score', 0)

    results.append({
        'username': author,
        'content': content,
        'upvotes': upvotes
    })

# ✅ Save to CSV
df = pd.DataFrame(results)
df.to_csv("reddit_ai_therapy_evilautism.csv", index=False, encoding='utf-8')
print(f"✅ Saved {len(df)} comments to reddit_ai_therapy_evilautism.csv")
```

```python
import pandas as pd

# ✅ Load both CSV files
df1 = pd.read_csv("reddit_ai_therapist_comments.csv")
df2 = pd.read_csv("reddit_ai_therapy_evilautism.csv")

# ✅ Combine into one DataFrame
combined = pd.concat([df1, df2], ignore_index=True)

# ✅ Drop duplicates based on content
cleaned = combined.drop_duplicates(subset="content")

# ✅ Save to new CSV
cleaned.to_csv("reddit_ai_therapy_combined.csv", index=False, encoding="utf-8")

print(f"✅ Done! Combined and saved {len(cleaned)} unique comments to reddit_ai_therapy_combined.csv")
```
